```python
import numpy as np
a = 2
rows_1=[]
for i in range(5):
    row = []
    for i in range(1, 6):
        row.append(int(np.power(a, i)))
    rows_1.append(row)
    a += 2
v = np.array([2.9, 14.8, 39.6, 74.3, 119])
x_1=np.linalg.solve(rows_1,v)
print(x_1)

rows_2=[]
for i in range(3):
    row = []
    for i in range(1, 4):
        row.append(int(np.power(a, i)))
    rows_2.append(row)
    a += 2
v = np.array([2.9, 14.8, 39.6,])
x_2=np.linalg.solve(rows_2,v)
print(x_2)
```

```python
import numpy as np
from sympy import Matrix
A_1=np.array([[8,-3,0,-7,2],
              [-9,4,5,11,-7],
              [6,-2,2,-4,4],
              [5,-1,7,0,10]])
A_2=np.array([[12,10,-6,-3,7,10],
              [-7,-6,4,7,-9,5],
              [9,9,-9,-5,5,-1],
              [-4,-3,1,6,-8,9],
              [8,7,-5,-9,11,-8]])
def rref_method(matrix):
    mat=Matrix(matrix)
    rrref_mat,pivot_cols=mat.rref()
    return np.array(matrix)[:,pivot_cols],pivot_cols
B_1,indices=rref_method(A_1)
B_2,indices=rref_method(A_2)
print("47B")
print(B_1)
print("48B")
print(B_2)
```

```python
import numpy as np
m=np.array([[0.95,0.03],[0.05,0.97]])
x_1=np.array([[600_000],[400_000]])
x_2=np.array([[350_000],[650_000]])
def count_population(x_0):
   x=x_0.copy()
   A=[]
   for i in range(20):
     x=m @ x
     A.append(x)
   return A
for k in range(1,21):
  a=count_population (x_1)
  city, suburb=a[k-1]
  print(f"k={k},city={int(city)},suburb={int(suburb)}")

for k in range(1,21):
  a=count_population (x_2)
  city, suburb=a[k-1]
  print(f"k={k},city={int(city)},suburb={int(suburb)}")


```
